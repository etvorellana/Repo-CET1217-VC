Como o cérebro percebe estímulos visuais? Quando consideramos como apresentar os dados da nossa pesquisa, devemos nos perguntar com que precisão o leitor consegue perceber os valores destes dados.

Com base em pesquisa desenvolvidas nos últimos trinta, [Schwabish](https://www.google.com.br/books/edition/Better_Data_Visualizations/A7jeDwAAQBAJ?hl=pt-BR&gbpv=0) elaborou a imagem a seguir que mostra um espectro de gráficos — ou mais genericamente, tipos de codificações de dados como pontos, linhas e barras — organizados com base na facilidade com que os leitores podem estimar seu valor. As codificações que os leitores podem estimar com mais precisão estão organizadas no topo, e aquelas que permitem estimativas mais gerais aparecem na parte inferior.

![Figura 1](Imagens/img_01.png)

Gráficos convencionais, como gráficos de barras e linhas, são tão comuns porque são perceptualmente mais precisos, familiares às pessoas e fáceis de criar. Gráficos menos comuns, como aqueles que usam círculos ou curvas, por exemplo, podem não permitir que o leitor perceba com mais precisão os valores exatos dos dados.

Mas a percepção da precisão nem sempre é o objetivo principal. E às vezes nem é um objetivo.

A visualização de dados é uma mistura de ciência e arte. Às vezes, queremos estar mais próximos do lado científico do espectro. Em outras palavras, usar visualizações que permitam aos leitores perceber com mais precisão os valores absolutos dos dados e fazer comparações. Outras vezes, podemos querer estar mais próximos do lado artístico do espectro e criar visuais que envolvam e entusiasmem o leitor, mesmo que não permitam as comparações mais precisas.

### 1.1 Quarteto de Anscombe

Em 1973 o estatístico [Francis Anscombe](https://www.jstor.org/stable/2682899) um artigo no qual ilustrou a importância da visualização dos dados utilizando o __Quarteto de Anscombe__. Neste [link](https://pt.wikipedia.org/wiki/Quarteto_de_Anscombe) da __Wikipedia__ podemos ter a tabela, mostrada na Figura a seguir, com quatro pares de dados $X$ e $Y$.  

![Figura 2](Imagens/img_02.png)

Podemos fazer algumas observações básicas sobre esses dados.
* Podemos ver que as três primeiras séries de $X$ são todas iguais; 
* Os valores de $X$ na última série são todos 8, exceto um que é 19; 
* Os $X$ são todos números inteiros, enquanto os $Y$ não são. 
* Podemos notar que o valor 12,7 na terceira coluna de $Y$ é maior que o resto. 

Acontece que cada um dos quatro pares, quando submetidos a um teste estatístico simples, produz a mesma informação: os mesmos valores médios da série $X$ e da série $Y$; a mesma variância para cada um; a mesma correlação entre $X$ e $Y$; e a mesma equação de regressão estimada. Veja na tabela a seguir, também obtida do [link](https://pt.wikipedia.org/wiki/Quarteto_de_Anscombe) da __Wikipedia__. 

![Figura 3](Imagens/img_03.png)

```python
#Anscombe Quartet
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import numpy as np
```

```python
X1 = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
Y1 = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])  #Anscombe's first dataset
X2 = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
Y2 = np.array([9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74])  #Anscombe's second dataset
X3 = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
Y3 = np.array([7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73])  #Anscombe's third dataset
X4 = np.array([8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8])
Y4 = np.array([6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89])  #Anscombe's fourth dataset
```

```python
print(f"Média de X1 = {np.mean(X1)}")
print(f"Média de X2 = {np.mean(X2)}")
print(f"Média de X3 = {np.mean(X3)}")
print(f"Média de X4 = {np.mean(X4)}")
```

```python
print(f"Média de Y1 = {np.mean(Y1)}")
print(f"Média de Y2 = {np.mean(Y2)}")
print(f"Média de Y3 = {np.mean(Y3)}")
print(f"Média de Y4 = {np.mean(Y4)}")
```

```python
print(f"Variância de X1 = {np.var(X1)}")
print(f"Variância de X2 = {np.var(X2)}")
print(f"Variância de X3 = {np.var(X3)}")
print(f"Variância de X4 = {np.var(X4)}")
```

```python
print(f"Variância de Y1 = {np.var(Y1)}")
print(f"Variância de Y2 = {np.var(Y2)}")
print(f"Variância de Y3 = {np.var(Y3)}")
print(f"Variância de Y4 = {np.var(Y4)}")
```

```python
print(f"Coeficiente de correlação de X1 e Y1 = {np.corrcoef(X1, Y1)[0,1]}")
print(f"Coeficiente de correlação de X2 e Y2 = {np.corrcoef(X2, Y2)[0,1]}") #Correlação de Pearson  entre X2 e Y2
print(f"Coeficiente de correlação de X3 e Y3 = {np.corrcoef(X3, Y3)[0,1]}") #Correlação de Pearson  entre X3 e Y3
print(f"Coeficiente de correlação de X4 e Y4 = {np.corrcoef(X4, Y4)[0,1]}") #Correlação de Pearson  entre X4 e Y4
```

```python
#Regressão linear
m1, b1 = np.polyfit(X1, Y1, 1)
m2, b2 = np.polyfit(X2, Y2, 1)
m3, b3 = np.polyfit(X3, Y3, 1)
m4, b4 = np.polyfit(X4, Y4, 1)
print(f"Coeficientes da regressão linear de X1 e Y1: y ={m1} x + {b1}")
print(f"Coeficientes da regressão linear de X2 e Y2: y ={m2} x + {b2}")
print(f"Coeficientes da regressão linear de X3 e Y3: y ={m3} x + {b3}")
print(f"Coeficientes da regressão linear de X4 e Y4: y ={m4} x + {b4}")
```

```python
#Coeficiente de determinação da regressão linear: R^2
r1 = np.corrcoef(X1, Y1)[0,1] ** 2
r2 = np.corrcoef(X2, Y2)[0,1] ** 2
r3 = np.corrcoef(X3, Y3)[0,1] ** 2
r4 = np.corrcoef(X4, Y4)[0,1] ** 2
print(f"Coeficiente de determinação da regressão linear de X1 e Y1: {r1}")
print(f"Coeficiente de determinação da regressão linear de X2 e Y2: {r2}")
print(f"Coeficiente de determinação da regressão linear de X3 e Y3: {r3}")
print(f"Coeficiente de determinação da regressão linear de X4 e Y4: {r4}")
```

Entretanto, quando vemos os mesmos dados apresentados em quatro gráficos, podemos ver imediatamente as relações, por exemplo: 
* a correlação positiva em todos os quatro pares; 
* a curvatura no segundo par que não é evidente na tabela;  
* os outliers 12,7 e 19,0. 

Temos muito mais probabilidade de lembrar desses quatro pequenos gráficos do que da tabela original. 

Em seu livro best-seller, Brain Rules, o biólogo molecular John Medina escreve: "Quanto mais visual a entrada se torna, mais provável é que seja reconhecida e lembrada". Quanto mais pudermos tornar nossos dados e conteúdo visuais, mais podemos esperar que nossos leitores se lembrem deles e, com sorte, os usem.

```python
#Gráficos do quarteto de Anscombe
fig, axs = plt.subplots(2, 2)
Xr = np.linspace(0, 20, 100)
fig.suptitle("Quarteto de Anscombe")
axs[0, 0].scatter(X1, Y1, color='r')
axs[0, 0].plot(Xr, m1*Xr + b1, color='b')
axs[0, 0].set_title('Dataset 1')
axs[0, 1].scatter(X2, Y2, color='g')
axs[0, 1].plot(Xr, m2*Xr + b2, color='b')
axs[0, 1].set_title('Dataset 2')
axs[1, 0].scatter(X3, Y3, color='b')
axs[1, 0].plot(Xr, m3*Xr + b3, color='b')
axs[1, 0].set_title('Dataset 3', y=-0.3)
axs[1, 1].scatter(X4, Y4, color='y')
axs[1, 1].plot(Xr, m4*Xr + b4, color='b')
axs[1, 1].set_title('Dataset 4', y=-0.3)
plt.show();
```

### 1.2 Princípios da Gestalt de Percepção Visual

Como percebemos informações? E como, como criadores de gráficos, podemos usar essas regras perceptivas para comunicar nossos dados de forma mais eficaz? A "teoria da Gestalt" é uma dessas maneiras de pensar sobre como nossos leitores olharão para nossos gráficos. A teoria da Gestalt foi desenvolvida no início do século XX por psicólogos alemães e se refere a como tendemos a organizar elementos visuais em grupos. Outros desenvolvimentos no campo foram interrompidos pela ascensão do regime nazista na Alemanha e depois pela Segunda Guerra Mundial, e depois da guerra foi criticado por não ter métodos metodológicos rigorosos. Mas as ideias persistem em muitas disciplinas, incluindo teoria da informação, ciência da visão e neurociência cognitiva.

Esses seis princípios organizacionais da teoria da Gestalt são especialmente úteis para criar gráficos e visuais que exploram a rede de processamento visual do nosso leitor.



#### 1.2.1 Proximidade

Percebemos objetos que estão próximos uns dos outros como pertencentes a um grupo. Há muitos elementos gráficos que podemos agrupar: rótulos com pontos, barras entre si ou, como no gráfico a seguir, grupos de pontos em um diagrama de dispersão no qual podemos ver dois grupos ou clusters, um no canto superior direito e o outro mais próximo do canto inferior esquerdo.

```python
# gerar um conjunto de pares X, Y com distribuição normal 
# em torno do ponto (3,3) com desvio padrão de 1.5
X1 = np.random.normal(3, 3.5, 300)
Y1 = np.random.normal(3, 1.5, 300)
# gerar um conjunto de pares X, Y com distribuição normal 
# em torno do ponto (17,16) com desvio padrão de 2.5
X2 = np.random.normal(12, 2.5, 300)
Y2 = np.random.normal(13, 2.5, 300)
# fazer um scater plot dos dois conjuntos de dados utilizando 
# a mesma cor para os dois conjuntos sem mostrar os eixos 
# coordenados e sem mostrar a legenda. Utilizar círculos
# sem preenchimento para os pontos
plt.scatter(X1, Y1, color='b', edgecolor='b', facecolor='none')
plt.scatter(X2, Y2, color='b', edgecolor='b', facecolor='none')
plt.axis('off')

plt.show();
```

#### 1.2.2 Semelhança ou similaridade

Nossos cérebros agrupam objetos que compartilham a mesma cor, forma ou direção. Adicionar cor ao gráfico de dispersão acima reforça os dois grupos.

```python
# gerar um conjunto de pares X, Y com distribuição normal 
# em torno do ponto (3,3) com desvio padrão de 1.5
X1 = np.random.normal(3, 3.5, 300)
Y1 = np.random.normal(3, 1.5, 300)
# gerar um conjunto de pares X, Y com distribuição normal 
# em torno do ponto (17,16) com desvio padrão de 2.5
X2 = np.random.normal(12, 2.5, 300)
Y2 = np.random.normal(13, 2.5, 300)
# fazer um scater plot dos dois conjuntos de dados utilizando 
# a mesma cor para os dois conjuntos sem mostrar os eixos 
# coordenados e sem mostrar a legenda. Utilizar círculos
# sem preenchimento para os pontos
plt.scatter(X1, Y1, color='r', edgecolor='r', facecolor='none')
plt.scatter(X2, Y2, color='b', edgecolor='b', facecolor='none')
plt.axis('off')

plt.show();
```

#### 1.2.3 Compartimento

Objetos delimitados são percebidos como um grupo. Aqui, além de usar cores, podemos delimitar os dois grupos com círculos ou outras formas geométricas.

```python
# gerar um conjunto de pares X, Y com distribuição normal 
# em torno do ponto (3,3) com desvio padrão de 1.5
X1 = np.random.normal(3, 3.5, 300)
Y1 = np.random.normal(3, 1.5, 300)
# gerar um conjunto de pares X, Y com distribuição normal 
# em torno do ponto (17,16) com desvio padrão de 2.5
X2 = np.random.normal(12, 2.5, 300)
Y2 = np.random.normal(13, 2.5, 300)
X1min = np.min(X1) - 1.0
X1max = np.max(X1) + 1.0
Y1min = np.min(Y1) - 1.0
Y1max = np.max(Y1) + 1.0
X2min = np.min(X2) - 1.0
X2max = np.max(X2) + 1.0
Y2min = np.min(Y2) - 1.0
Y2max = np.max(Y2) + 1.0

# fazer um scater plot dos dois conjuntos de dados utilizando 
# a mesma cor para os dois conjuntos sem mostrar os eixos 
# coordenados e sem mostrar a legenda. Utilizar círculos
# sem preenchimento para os pontos
elipse1 = Ellipse((X1.mean(), Y1.mean()), 
                   X1max - X1min, 
                   Y1max - Y1min, edgecolor='r', facecolor='none')
elipse2 = Ellipse((X2.mean(), Y2.mean()),
                     X2max - X2min,
                     Y2max - Y2min, edgecolor='b', facecolor='none')
plt.gca().add_artist(elipse1)
plt.gca().add_artist(elipse2)

plt.scatter(X1, Y1, color='r', edgecolor='r', facecolor='none')
plt.scatter(X2, Y2, color='b', edgecolor='b', facecolor='none')
plt.axis('off')
# desenhar uma elipse em torno do primeiro conjunto de dados
# com diâmetros a X1max - X1min e Y1max - Y1min
# desenhar uma elipse em torno do segundo conjunto de dados
# com diâmetros a X2max - X2min e Y2max - Y2min


plt.show();
```

#### 1.2.6 Conexão

De acordo com esse princípio, percebemos objetos conectados como membros do mesmo grupo. Veja esta série de pontos: a princípio, percebemos como uma única série, uma massa de pontos azuis. Adicionar cor deixa claro que há duas séries diferentes. Conectar os pontos deixa claro como os dois inicialmente se acompanham, mas depois divergem.

```python
X = np.linspace(0, 50, 51)
Y1 = np.zeros_like(X)
Y2 = np.zeros_like(X)
Y1[0:10] = np.cos(X[0:10]*np.pi/5)*3 + 3
Y2[0:10] = np.sin(X[0:10]*np.pi/5) + 1
Y1[10:25] = np.cos(X[10:25]*np.pi/5) + 1
Y2[10:25] = np.sin(X[10:25]*np.pi/5) + 1
Y2[25:] = np.sin(X[25:]*np.pi/5)*0.5 + X[25:]/5
Y1[25:] = np.cos(X[25:]*np.pi/5)*0.8 + X[25:]/90
print(X.shape, Y1.shape, Y2.shape)
fig, ax = plt.subplots(3, 1)
ax[0].scatter(X, Y1, color='b', s=5)
ax[0].scatter(X, Y2, color='b', s=5)
ax[1].scatter(X, Y1, color='r', s=5)
ax[1].scatter(X, Y2, color='b', s=5)
ax[2].scatter(X, Y1, color='r', s=5)
ax[2].scatter(X, Y2, color='b', s=5)
ax[2].plot(X, Y1, 'r-')
ax[2].plot(X, Y2, 'b-')
```
